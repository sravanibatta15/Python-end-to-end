**Do Python Tuple is Mutable or Immutable -> Prove it**

- `Try to Increase the Memory `
- `Try to Decrease the Memory `

```python
a = (10,50,20) # save in the Memory

a[3] = 100

print(a) # calling
```

```python
a = (10,50,20) # save in the Memory

del a[1]

print(a) # calling
```

- From the Above operations we can say that tuple is immutable

```python
# tuple having only 2 Built in Functions

# count
# index
```

```python
a = (10,50,20) # save in the Memory

print(a.count(50))
```

```python
print(a.index(20))
```
