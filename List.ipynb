**Built_in_functions In list**

Adding Memory
- `append `
- `insert `
- `extend `

Decreasing Memory
- `pop `
- `remove `
- `clear `

For some Common Operations
- `count `
- `index `
- `sort `
- `reverse`
- `copy `


```python
# append = will take single input from user and will add to the end of the existing List

a = [10,20,50,4] # save in the Memory

a.append(100) # no need of varibale updation because list automaticaaly it updates

print(a) # calling
```

```python
# extend = will take multiple inputs from user and will add to the end of the existing list

a = [10,20,50,4] # save in the Memory

a.extend([100,200,300])

print(a) # calling
```

```python
# Insert = we can add values where we want

a = [10,20,50,4] # save in the Memory

a.insert(2 , 100)

print(a) # calling
```

```python
# pop = defautly will remove last value
# If user passes any input it wil consider as index

a = [10,20,50,4] # save in the Memory

a.pop()

print(a)
```

```python
# pop = defautly will remove last value
# If user passes any input it wil consider as index

a = [10,20,50,4] # save in the Memory

a.pop(2)

print(a)
```

```python
# remove

a = [10,20,50,4] # save in the Memory

a.remove(20)

print(a)
```

```python
a = [10,20,50,4] # save in the Memory

del a # entire varibale is deleted


print(a) # calling
```

```python
a = [10,20,50,4] # save in the Memory

a.clear() # will remove all values in the varibae but varibale will be alive

print(a) # calling
```

- `Python List is Mutable -> Increasing and decreasing the Memory is possible`

```python
# count = used to count the number of occurences

a = [10,20,30,50,10,90]

print(a.count(50))
```

```python
# Index = used to find the Index of a value

a = [10,20,30,50,10,90]

print(a.index(10 , 1))
```

```python
a = [10,20,30,50,10,90]

a[3] = 500

print(a)
```

```python
# reverse -> [::-1]

a = [10,20,30,50,10,90]

a.reverse() # [::-1]

print(a) # calling
```

```python
# sort

a = [10,5,1,9,11,2,16,7]


a.sort() # defaulty will apply ascending order

print(a)
```

```python
# sort

a = [10,5,1,9,11,2,16,7]


a.sort(reverse = True) # defaulty will apply ascending order

print(a)
```

```python
# copy

a = [10,20,30] # save in the Memory
b = a # deep copy

print(a)
print(b)

print(id(a))
print(id(b))



a.append(100)


print(a)
print(b)
```

```python
# copy

a = [10,20,30] # save in the Memory
b = a.copy() # shallow copy

print(a)
print(b)

print(id(a))
print(id(b))



a.append(100)


print(a)
print(b)
```
