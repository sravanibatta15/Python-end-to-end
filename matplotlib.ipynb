**Matplotlib**

Matplotlib is a EDA library

- `scatter plot `
- `line chart `
- `legend function `
- `subplots `
- `histogram `
- `bar chart `
- `pie chart `
- `image data `

```python
import matplotlib.pyplot as plt
import numpy as np
```

**scatter_plot**

```python
india = [1,5,7,3,6]
usa = [5,7,1,6,3]

plt.figure(figsize = (5,3))

plt.scatter(x = india,y = usa) # math will happen

plt.show()
```

```python
india = [1,5,7,3,6]
usa = [5,7,1,6,3]

plt.figure(figsize = (5,3))


plt.title('Scatter_plot')
plt.xlabel('india_values')
plt.ylabel('usa_values')

plt.scatter(x = india,y = usa , color = 'r' , marker = 'd') # math will happen

plt.show()
```

```python
# line chart

india = [1,5,7,3,6]
usa = [5,7,1,6,3]

plt.figure(figsize = (5,3))


plt.title('line_chart')
plt.xlabel('india_values')
plt.ylabel('usa_values')

plt.plot(india,usa,color='r',marker='*')

plt.show()
```

```python
# line chart

india = [1,5,7,3,6]
usa = [5,7,1,6,3]
canada = [3,7,1,6,4]

plt.figure(figsize = (5,3))


plt.title('line_chart')
plt.xlabel('india_values')
plt.ylabel('usa_canada_values')

plt.plot(india,usa,color='r',marker='*')
plt.plot(india,canada,color='g',marker='s')

plt.show()
```

```python
# legend

india = [1,5,7,3,6]
usa = [5,7,1,6,3]
canada = [3,7,1,6,4]

plt.figure(figsize = (5,3))


plt.xlabel('india_values')
plt.ylabel('usa_canada_values')

plt.plot(india,usa,color='r',marker='*',label='india-usa')
plt.plot(india,canada,color='g',marker='s',label='india-canada')

plt.legend()

plt.show()
```

**subplots**

```python
india = [1,5,7,3,6]
usa = [5,7,1,6,3]
canada = [3,7,1,6,4]

plt.figure(figsize = (8,3))


plt.subplot(1,2,1)
plt.title('india_usa')
plt.xlabel('india')
plt.ylabel('usa')
plt.plot(india,usa,color='r',marker='*')


plt.subplot(1,2,2)
plt.title('india_canada')
plt.xlabel('india')
plt.ylabel('canada')
plt.plot(india,canada,color='g',marker='s')

plt.show()
```

```python
india = [1,5,7,3,6]
usa = [5,7,1,6,3]
canada = [3,7,1,6,4]

plt.figure(figsize = (5,5))


plt.subplot(2,1,1)
plt.title('india_usa')
plt.xlabel('india')
plt.ylabel('usa')
plt.plot(india,usa,color='r',marker='*')


plt.subplot(2,1,2)
plt.title('india_canada')
plt.xlabel('india')
plt.ylabel('canada')
plt.plot(india,canada,color='g',marker='s')

plt.show()
```

```python
# histogram = 1D
# which is used to find the data in between the intervals

age = [1,3,4,7,9,11,14,16,18,22,29,27,33,35]

plt.figure(figsize = (5,3))

plt.hist(x = age,bins = 4)

plt.show()
```

```python
# Bar chart

company = ['google','msft','meta','ibm']
revenue = [1600,800,1100,1900]

plt.figure(figsize = (5,3))

plt.bar(company,revenue)

plt.show()
```

```python
# Bar chart

company = ['google','msft','meta','ibm']
revenue = [1600,800,1100,1900]

plt.figure(figsize = (5,3))

plt.bar(company,revenue , color = ['r','g','black','y'])

plt.show()
```

```python
# Bar chart

company = ['google','msft','meta','ibm']
revenue = [1600,800,1100,1900]

plt.figure(figsize = (5,3))

plt.barh(company,revenue , color = ['r','g','black','y'])

plt.show()
```

```python
# pie chart

company = ['google','msft','meta','ibm']
revenue = [1600,800,1100,1900]

plt.figure(figsize = (5,3))

plt.pie(revenue,labels = company)

plt.show()
```

```python
# pie chart

company = ['google','msft','meta','ibm']
revenue = [1600,800,1100,1900]

plt.figure(figsize = (5,3))

plt.pie(revenue,labels = company,shadow=True)

plt.show()
```

```python
# pie chart

company = ['google','msft','meta','ibm']
revenue = [1600,800,1100,1900]

plt.figure(figsize = (5,3))

plt.pie(revenue,labels = company,shadow=True , autopct = '%.2f')

plt.show()
```
