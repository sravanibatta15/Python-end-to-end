```python
# Given Number is a Palindrome or not:


a = 121
if int(str(a)[::-1]) == a:
  print('Palindrome')
else:
  print('Not')
```

```python
a = 121
temp = a
s = 0


while a > 0:
  res = a % 10
  s = (s * 10) + res
  a = a // 10

if temp == s:
  print('Palindrome')
else:
  print('not a palindrome')
```

```python
# prove the given number is amstrong or not

a = 153
temp = a
s = 0 # 152


while a > 0:
  res = a % 10 # 1
  s = s + (res * res * res)
  a = a // 10 # 0

if temp == s:
  print('Amstrong')
else:
  print('not a Amstrong')
```

```python
# Fibonacci Series

n = 8
a = 0
b = 1
print(a)
print(b)

for i in range(6):
  c = a + b
  print(c)
  a = b
  b = c
```
