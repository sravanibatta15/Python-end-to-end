```python
# 1D

import numpy as np

a = np.array([10,20,30,50,90])

print(a)

print(a.ndim)

print(a.shape)

print(a.size)
```

```python
a = np.array([10,20,30,50,90])

print(a)

print(a[2])
```

```python
a = np.array([  [10,90,60] , [55,44,76] , [3,17,14] ])

print(a)

print(a.ndim)

print(a.shape)

print(a.size)
```

```python
a = np.array([  [10,90,60] , [55,44,76] , [3,17,14] ])

print(a)

print(a[1][2])
```

```python
# arange = is a built in function in numpy which is used to create data in
# 1D this function will take starting point and ending point - 1

a = np.arange(21,27)
print(a)

print(a.ndim)
```

```python
a = np.arange(55,66)
print(a)
print(a.ndim)
```

```python
a = np.arange(1,61)
print(a)
print(a.ndim)

b = a.reshape(10,6)
print(b)
print(b.shape)
```

```python
a = np.arange(1,61).reshape(10,6)
print(a)
print(a.shape)
```

```python
print(a[4][3])
```

```python
# 1d to 3D

a = np.arange(1,61).reshape(5,2,6)
print(a)
```

```python
a[2][0][3]
```

```python
# 1d to 3D

a = np.arange(1,61).reshape(2,6,5)
print(a)
```

```python
a[0][3][3]
```

```python
a[1][1][2]
```

```python
# reshape(-1,1) and reshape(1,-1)

a = np.array([10,50,90,11,14])
print(a)
print(a.ndim)
print(a.shape)


b = a.reshape(-1,1)
print(b)
print(b.ndim)
print(b.shape)
```

```python
# reshape(-1,1) and reshape(1,-1)

a = np.array([10,50,90,11,14])
print(a)
print(a.ndim)
print(a.shape)


b = a.reshape(1,-1)
print(b)
print(b.ndim)
print(b.shape)
```

```python
# linespace

a = np.linspace(10,20,6)
print(a)
```

```python
# zeros

a = np.zeros((4,5))
print(a)

print(a.dtype)
```

```python
# zeros

a = np.zeros((4,5) , dtype=np.float16)
print(a)

print(a.dtype)
```

```python
# zeros

a = np.zeros((4,5) , dtype=np.int8)
print(a)

print(a.dtype)
```

```python
# ones

a = np.ones((5,6) , dtype=np.int8)
print(a)
```

```python
# full

a = np.full((4,5) , 7)
print(a)
```

```python
# eye

a = np.eye(4)
print(a)
```

```python
# ravel = any dim into 1D

a = np.array([  [10,90,60] , [55,44,76] , [3,17,14] ])

print(a)

b = a.ravel()
print(b)
```

```python
# concatinate

a = np.array([  [10,90,60] , [55,44,76] , [3,17,14] ])
b = np.array([ [1,2,3] , [5,6,7] , [10,11,12] ])

print(a)
print(b)

c = np.concatenate((a,b))
print(c)
```

```python
# Transpose = rows to columns and columns to rows

a = np.array([  [10,90,60] , [55,44,76] , [3,17,14] ])
print(a)

print()


b = np.transpose(a) # a.T
print(b)
```

```python
# random -> randint = between range randomly int values will be picked

a = np.random.randint(10,50,5)

print(a)
```

```python
# random -> rand = based on rows and columns 0 to 1 random values will come

a = np.random.rand(5,6)
print(a)
```

```python
# random -> randn = based on rows and columns -3 to 3 random values will come

a = np.random.randn(5,6)
print(a)
```

```python
# random -> random = same like rand will give values between 0 to 1

a = np.random.random((5,6))
print(a)
```

```python
# repeat

a = np.arange(50,61)
print(a)
print(a.shape)
print(a.ndim)


b = a.reshape(1,-1)
print(b)
print(b.shape)
print(b.ndim)
```

```python
# repeat

a = np.arange(50,61)

b = a.reshape(1,-1) # 1d to 2d

c = np.repeat(b , 10 , axis = 0)
print(c)
```

```python
# loop for numpy

f = np.array([ [2,1,10] , [11,14,7] , [10,20,1]])

p = []

for i in f:
  t = 0
  for j in i:
    t = t + j

  p.append(t)

print(p)
```
