**What is Global varibales and Local Varibales**

```python
# Global varibale means outside the Function
# Local variables means inside the Function
```

```python
f = 100 # Global variable

def kiran():
  print(f"I am Inside the Function : {f}")


kiran() # calling

print(f"I am Outside the Function : {f}")
```

```python
def kumar():
  t = 500   # local variable
  print(f'I am from Inside the Function : {t}')


kumar() # calling
print(f'I am from Outside the Function : {t}')
```

```python
def kumar():
  global t
  t = 500   # local variable to global varible
  print(f'I am from Inside the Function : {t}')


kumar() # calling
print(f'I am from Outside the Function : {t}')
```
