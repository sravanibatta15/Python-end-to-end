```python
class perfect:
  def __init__(self,n):
    self.n = n
    self.c = 0

  def calculation(self):
    for i in range(1,self.n):
      if self.n % i == 0:
        self.c = self.c + i

    if self.c == self.n:
      print("Perfect")
    else:
      print('Not A perfect ')

obj = perfect(6)
obj.calculation()
```

```python
import sys

class PALINDROME:
  try:
    def __init__(self , f):
      self.f = f
  except Exception as e:
    e_t,e_m,e_lin = sys.exc_info()
    print(f'Error is From Line No : {e_lin.tb_lineno} Because {e_t}:')


  def calculation(self):
    try:
      for i in self.f:
        s = 0
        t = i
        while i > 0:
          res = i % 10
          s = (s*10)+ res
          i = i // 10
        if t == s:
          with open('result.txt','a') as f:
            f.write(f" it is a Palindrome : {t}")
        else:
          with open('result.txt','a') as f:
            f.write(f' it is Not a Palindrome : {t}')

    except Exception as e:
      e_t,e_m,e_lin = sys.exc_info()
      print(f'Error is From Line No : {e_lin.tb_lineno} Because {e_t}:')



kiran = PALINDROME([135,121,145])
kiran.calculation()
```

**Inheritance**

Inheritance is like a relation between parent and child  

There are 5 types of Inheritance

- `Single Inheritance `
- `Multiple Inheritance `
- `Multi Level Inheritance `
- `hierarical Inhertiance `
- `Hybrid Inheritance `

```python
# single Inheritance

class PARENT:
  def building(self):
    print('I have My own Building')



class CHILD(PARENT):
  def job(self):
    print('I have My own Job')



anil = PARENT()
anil.building()
```

**PolyMorphisam**

- `Poly -> many`
- `Morphisam -> forms `


In polymorphisam we have 2 Techniques

- `Method [Functions] Overloading `
- `Method [Functions] Overriding `

```python
# method overriding

def fun():
  print('Good Morning')


def fun():
  print('Good Evening')


fun() # calling
```

```python
# Method Overriding = is a technique which will be used
# only when we have one parent and one child class


class MANAGER:
  def database_connection(self):
    print("I am connected to MySQL DataBase")


class MODULE_LEAD(MANAGER):
  def database_connection(self):
    print('Code changed successfully ')


kiran = MODULE_LEAD()
kiran.database_connection()
```

```python
# Method OverLoading

def loading(a = None,b = None):
  if a == None and b == None:
    print('I am from USA')

  elif a == None and b != None:
    print('I am from UK')

  elif a != None and b == None:
    print('I am from canada')

  else:
    print('I am from India')


loading()
loading(10)
loading(b= 20)
loading(a=10,b=20)
```

```python
# Abstraction

from abc import ABC,abstractmethod


class second_version(ABC):
  @abstractmethod
  def UPI_Transaction(self):
    pass


class Version_3(second_version):
  def UPI_Transaction(self):
    print('I am Implementing')

kiran = Version_3()
kiran.UPI_Transaction()
```
