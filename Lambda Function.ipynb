**Lambda Function**

`Lambda Functions are also a functions but they dont have Function Name`

- `syntax :    lambda arguments : expressions `

```python
a = 10
b = 20

print(a + b)
```

```python
# normal function

def fun(a,b):
  print(a+b)

fun(10,20) # calling
```

```python
c = lambda a,b : a+b

print(c(10,20))
```

```python
n = 2

for i in range(1,11):
  print(n,'*',i,'=',n*i)
```

```python
def sharuk(n):
  for i in range(1,11):
    print(n,'*',i,'=',n*i)



sharuk(2) # calling
```

```python
sharuk(3)
```

```python
c = lambda x,y : x*y
n = 2



for i in range(1,11):
  print(n,'*',i,'=',c(n,i))
```

**Map | Filter | Reduce**

```python
# Map function is used to take values one after other
# if we use Map Function there is no need of for loop
# syntax: map(function_name , iterables)
```

```python
def sharuk(b):
  if b % 2 == 0:
    return b



sol = map(sharuk , [10,25,35,40,55,60])

print(list(sol))
```

```python

sol = map(lambda x : x + 20, [10,25,35,40,55,60])

print(list(sol))
```

```python
def sharuk(b):
  if b % 2 == 0:
    return b



sol = filter(sharuk , [10,25,35,40,55,60])

print(list(sol))
```

```python
import functools

print(functools.reduce(lambda x,y : x + y , [1,4,7,9,3,2]))
```

**Map with Filter**

```python
a=map(lambda x:x+20 , filter(lambda x:x%2==0,[1,2,3,9,11,14,12]))
print(list(a))
```

```python
# filter with Map

a=filter(lambda x:x%2==0 , map(lambda x:x+20,[1,2,3,9,11,14,12]))
print(list(a))
```

```python
# reduce with map and filter

a=functools.reduce(lambda x,y:x+y,map(lambda x:x+20,filter(lambda x:x%2==0,[1,3,4,6,2,11,14,19,18])))
print(a)
```
