**Exceptional Handling Technique**

```python
a = 2
b = 10
c = b / a
print(c)
```

```python
try:
  a = 2
  b = 10
  c = b / a
  print(c)

except:
  pass
```

```python
a = 10
b = 0
print(a / b)
```

```python
try:
  a = 10 # save in the Memory
  b = 0 # save in the Memory
  print(a / b) # error in 8 lines -> will be give to try keyword

except Exception as e:
  print(e)
```

```python
try:
  a = 10 # save in the Memory
  b = 2 # save in the Memory
  print(a / b) # will work

except Exception as e:
  print(e)
```

```python
try:
  city = 'hyd'
  print(CITY)

except ZeroDivisionError as e:
  print(e)

except ValueError as e:
  print(e)
```

```python
try:
  city = 'hyd'
  print(CITY) # 8 lines

except Exception as e:
  print(e)
```

```python
# finally is a function in exceptional handling
# which will work with exception and without exceptional also
```

```python
try:
  city = 'hyd' # SAVE IN THE MEMORY
  print(city) # 8 lines

except Exception as e:
  print(e)


finally:
  print('The code is working')
```

```python
import sys

try:
  a = 10 # save in the Memory
  b = 2 # save in the Memory
  print(a / b) # error in 8 lines -> will be give to try keyword

except Exception as e:
  print(f'Error Msg Given By Try Block : {e}')
  e_type,e_msg,e_linno = sys.exc_info()
  print(f'Error Type : {e_type}')
  print(f'Error Msg by sys library: {e_msg}')
  print(f'Error Line no : {e_linno.tb_lineno}')
```
