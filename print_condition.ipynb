**We can create the relationship b/w data types inside the Print condition**

- `1. Using comma (,) -> comma will help us in creating the relation b/w 2 data types with one tab space `

- `2. Using (+) -> we can create relation b/w data types but when we are using + both should be in the form of str only `

- `3. Identifiers -> %d -> int | %f -> float | %s str `

- `4. Using {} and f `

```python
title = 'data science'
time = 6
class_ = 1.30


# I am learning data science in vihara tech for 6 months with 1.30 hours class daily

print('I am learning %s in vihara tech for %d months with %f hours class daily'%(title,time,class_))
```

```python
title = 'data science'
time = 6
class_ = 1.30


# I am learning data science in vihara tech for 6 months with 1.30 hours class daily

print('I am learning %s in vihara tech for %d months with %f hours class daily'%(time,title,class_))
```
