```python
# Mutable = If a varibale in the memory can change size means it is mutable
# Immutabe = If a varibale in the memory can't change size means it is Immutable
```

- `To Prove whether string is Mutable or Immutable -> we need to work on `

       - `Try to Increase the Memory size`
       - `Try to decrease the memory size `

```python
# trying to increase the Memory size

a = 'data'  # save in the Memory

a[4] = 'j'  # trying to add j in new box

print(a) # calling
```

```python
# trying to Decrease the Memory size

a = 'data'  # save in the Memory

del a[2]  # trying to add j in new box

print(a) # calling
```

```python
a='data'
a=a + 'j'
print(a)
```

**Since from the Above Operations we can say that String in Python is Immutable**
