**Python Dictionary**


- `Python Dict is going to represent in {} -> In a key_value pair `

- `where key means column name and value means data inside the column `

- `Dictionary is used to Prepare the data just like exsheet | but we can see a table | if we want to see please give DIct to Pandas Library `

- `we can pass anything in the place of values `

- `In the place of keys we can't pass mutable data types |`

- `in dict we dont have index | since there is no Index slicing and skipping is not possible `

**Built_in_functions in dictionary**

- `get`
- `update`
- `pop `
- `popitem `
- `keys `
- `values `
- `items `
- `clear `
- `copy `

```python
fruits = {'apple':50 , 'banana':20,'carrot':60}

print(fruits['banana'])
```

```python
fruits = {'apple':50 , 'banana':20,'carrot':60}

print(fruits.get('banana'))
```

```python
fruits = {'apple':50 , 'banana':20,'carrot':60}

fruits['grapes'] = 1000

print(fruits)
```

```python
fruits = {'apple':50 , 'banana':20,'carrot':60}

fruits['apple'][1] = 1000
print(fruits)
```

```python
fruits = {'apple':50 , 'banana':20,'carrot':60}

fruits.update({'grapes':1000})

print(fruits)
```

```python
fruits = {'apple':50 , 'banana':20,'carrot':60}

del fruits['banana']

print(fruits)
```

```python
fruits = {'apple':50 , 'banana':20,'carrot':60}

fruits.pop('banana')

print(fruits)
```

```python
fruits = {'apple':50 , 'banana':20,'carrot':60}

fruits.pop()

print(fruits)
```

```python
# popitem  = will remove last column

fruits = {'apple':50 , 'banana':20,'carrot':60}

fruits.popitem()

print(fruits)
```

```python
fruits = {'apple':50 , 'banana':20,'carrot':60}
b = fruits #deep copy

print(fruits)
print(b)

print(id(fruits))
print(id(b))

fruits['apple'] = 100

print(fruits)
print(b)
```

```python
fruits = {'apple':50 , 'banana':20,'carrot':60}
b = fruits.copy() # shallow copy

print(fruits)
print(b)

print(id(fruits))
print(id(b))

fruits['apple'] = 100

print(fruits)
print(b)
```
