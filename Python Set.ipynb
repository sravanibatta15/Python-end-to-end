```python
# add = add function will take a single input from user and will add to existing set randomlly

a = {10,20,30,50} # save in the Memory

print(f"Before : {a}")


a.add(100)

print(f'After : {a}')
```

```python
# update = will take multiple values and will add randomlly

a = {10,20,30,50} # save in the Memory

print(f"Before : {a}")

a.update([100,200,300])

print(f'After : {a}')
```

```python
# pop = will remove a value randomlly

a = {10,20,30,50} # save in the Memory

print(f"Before : {a}")

a.pop()

print(f'After : {a}')
```

```python
# pop = will remove a value randomlly

a = {10,20,30,50} # save in the Memory

print(f"Before : {a}")

a.pop(1)

print(f'After : {a}')
```

```python
# discard= will take input from user
# and if the value is there it will remove or else it will be silent

a = {10,20,30,50} # save in the Memory

print(f"Before : {a}")

a.discard(20)

print(f'After : {a}')
```

**From the Above operations we can say python set is mutable**

```python
a = {10,20,30}
b = a #deep copy

print(a)
print(b)

print(id(a))
print(id(b))

a.add(100)


print(a)
print(b)
```

```python
a = {10,20,30}
b = a.copy() # shallow copy

print(a)
print(b)

print(id(a))
print(id(b))

a.add(100)


print(a)
print(b)
```

```python
a = {1,2,3,4,5,6}
b = {5,6,7,8,9,10}

print(a.union(b))
```

```python
a = {1,2,3,4,5,6}
b = {5,6,7,8,9,10}

print(a.intersection(b))
```
