**File Handling**

- `Means we are going to work on files [.txt , .xlsx , .json , .csv]`


- `In File Handling concept Mainly we have 3 Modes `

  - `w -> write mode `
  - `r -> read mode `
  - `a -> add mode `

**Mode W**

- `Rule:1 -> If there is no file mode w will create the file and will write the data into the file `

- `Rule:2 -> If we have the file and there is no data in the file mode w will write data into it `
- `Rule:3 -> If we have file and If we have data in the file and if we pass a new data the existing data is going to override `

```python
a = open('sharuk.txt','w')

a.write('I am from India')

a.close()
```

```python
a = open('kiran.txt','w')

a.write('I am from paris')

a.close()
```

```python
a = open('city.txt','w')

a.write('I am from paris')

a.close()
```

**Mode_a**

```python
a = open('sharuk.txt','a')

a.write('I am from India')

a.close()
```

```python
a = open('kiran.txt','a')

a.write('I am from paris')

a.close()
```

```python
a = open('city.txt','a')

a.write(' I am from paris')

a.close()
```

**read -> r**

```python
a = open('city.txt','r')

print(a.read())

a.close()
```
