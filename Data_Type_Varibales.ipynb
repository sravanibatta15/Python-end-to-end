**Rules to create a varibale**

 - `Rule_no:1 -> A varibale should not start with a Number -> If you want to use please use Underscore `

 - `Rule_no:2 -> A single character or single word you can use as a varibale but Multiple words not possible as a varibale -> if we want to use please use underscore `

 - `Rule_no:3 -> Python Varibales are case sensitive -> capital letters and small letters`

 - `Rule_no:4 -> Dont use this Punctuations as a varibale -> !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~`

**Type_conversions**

Changing once data type into other data type

- `int -> float and int -> str `
- `float -> int and float -> str `
- `str -> int and str -> float `

```python
# int -> float

a = 12 # save in the Memory
print(a) # calling from memory
print(type(a)) # checking data type

b = float(a) # data type conversion
print(b) # calling from memory
print(type(b)) # checking data type
```

```python
a = 12
b = str(a)
print(b)
```

```python
# int -> str

a = 12 # save in the Memory
print(a) # calling from memory
print(type(a)) # checking data type

b = str(a) # data type conversion
print(b) # calling from memory
print(type(b)) # checking data type
```

```python
# float -> int

a = 12.22 # save in the Memory
print(a) # calling from memory
print(type(a)) # checking data type

b = int(a) # data type conversion
print(b) # calling from memory
print(type(b)) # checking data type
```

```python
# float -> str

a = 12.22 # save in the Memory
print(a) # calling from memory
print(type(a)) # checking data type

b = str(a) # data type conversion
print(b) # calling from memory
print(type(b)) # checking data type
```
