`Comprehension concept in python are used to write the code in single line `

`There are 2 types of comprehensions `

        - `List Comprehensions `
        - `Dictionary Comprehensions`

Comprehension Syntax:

- `1. If we have one forloop and one print condition `

    [print_area [forloop]]

- `2. If we have one forloop and one if and one print condition `

    [print_area     [forloop]  [if condition]]



- `3. If we have one for loop one if and one else means `

    [[print_area] [ifcondition] else [print_area]   [ forloop]]

**List Comprehension**

```python
# Example:1
for i in range(1,11):
  print('Good Morning' , end=' ')
```

```python
print(['Good Morning' for i in range(1,11)])
```

```python
# Example:2

for i in range(11,17):
  if i % 2 == 0:
    print(f'Even : {i}')
```

```python
print([f'Even : {i}' for i in range(11,17) if i % 2 == 0])
```

```python
# Example:3

for i in range(11,18):
  if i % 2 == 0:
    print(f'Even : {i}')
  else:
    print(f'Odd : {i}')
```

```python
print([f'Even : {i}' if i % 2 == 0 else f'Odd : {i}' for i in range(11,18)])
```

**Dictionary Comprehension**

```python
# Example: 1

fruits = {'apple':100,'Banana':200,'Carrot':300}

for i in fruits:
  fruits[i] = fruits[i] * 2

print(fruits)
```

```python
fruits = {'apple':100,'Banana':200,'Carrot':300}

print({i: fruits[i] * 2 for i in fruits})
```

```python
# Example: 2
fruits = {'apple':100,'Banana':200,'Carrot':300}

d = {}

for i in fruits:
  if fruits[i] > 100:
    d[i] = fruits[i]

print(d)
```

```python
fruits = {'apple':100,'Banana':200,'Carrot':300}

print({i:fruits[i] for i in fruits if fruits[i] > 100})
```

```python
# Example: 3
fruits = {'apple':90,'Banana':200,'Carrot':300}
d = {}

for i in fruits:
  if fruits[i] > 100:
    d[i] = fruits[i] * 2
  else:
    d[i] = fruits[i] * 3

print(d)
```

```python
fruits = {'apple':90,'Banana':200,'Carrot':300}


print({i:(j * 2 if j > 100 else j*3) for i,j in fruits.items()})
```

```python
import time
starting_time = time.time()

fruits = {'apple':90,'Banana':200,'Carrot':300}
d = {}

for i in fruits:
  if fruits[i] > 100:
    d[i] = fruits[i] * 2
  else:
    d[i] = fruits[i] * 3

print(d)


print(f"Total time to Execute : {time.time() - starting_time}")
```

```python
import time
starting_time = time.time()

fruits = {'apple':90,'Banana':200,'Carrot':300}

print({i:(j * 2 if j > 100 else j*3) for i,j in fruits.items()})

print(f"Total time to Execute : {time.time() - starting_time}")
```
